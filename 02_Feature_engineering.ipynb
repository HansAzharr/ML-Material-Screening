## Derived Host Material Features

```python
# Define the path to PubChem CSV
# Data is retrieved from National Center for Biotechnology Information (2025).
# https://pubchem.ncbi.nlm.nih.gov/periodic-table/

pubchem_csv_filepath = '/content/drive/MyDrive/Colab Notebooks/Masters Research Project/Data/PubChemElements_all.csv'

property_column_map = {
    'atomic_number': 'AtomicNumber',
    'atomic_mass': 'AtomicMass',
    'electronegativity': 'Electronegativity',
    'covalent_radius': 'AtomicRadius' # Using AtomicRadius as a proxy here
}
```

```python
element_properties = {}

try:
    # Load the CSV file, select relevant columns, and set 'Symbol' as index
    required_cols = ['Symbol'] + list(property_column_map.values())
    df_elements = pd.read_csv(pubchem_csv_filepath, usecols=required_cols, index_col='Symbol')

    # Convert relevant columns to numeric, coercing errors to NaN
    for csv_col in property_column_map.values():
         df_elements[csv_col] = pd.to_numeric(df_elements[csv_col], errors='coerce')

    # Convert the DataFrame to a dictionary of dictionaries and map column names
    element_properties = df_elements.rename(columns={v: k for k, v in property_column_map.items()}).to_dict('index')

    print(f"Successfully loaded properties for {len(element_properties)} elements from '{pubchem_csv_filepath}'.")
    print("Note: 'AtomicRadius' from the CSV is being used as the 'covalent_radius' feature.")

except FileNotFoundError:
    print(f"Error: The PubChem CSV file was not found at '{pubchem_csv_filepath}'. Please check the path and ensure it's mounted.")
    element_properties = {}
    print("Element properties dictionary is empty due to file not found. Subsequent feature engineering will be affected.")
except Exception as e:
    print(f"An error occurred while processing the PubChem CSV: {e}")
    element_properties = {}
    print("Element properties dictionary is empty due to an error during CSV processing. Subsequent feature engineering will be affected.")
```

```python
# Derived Host Material Features (14 features)

if 'elements' in df.columns:
    # Safely convert string representation of lists to actual lists
    # Use literal_eval for safer evaluation if the format is consistent with Python literals
    df['elements_parsed'] = df['elements'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

    # Function to get properties and calculate aggregations for a list of elements
    def calculate_elemental_aggregations(element_list):
        if not isinstance(element_list, list):
            return pd.Series([np.nan] * (4 * 3 + 1)) # Return Series with NaN for all 13+1 features if input is not a list

        results = {}
        props_dict = {}
        for prop_name in ['atomic_number', 'atomic_mass', 'electronegativity', 'covalent_radius']:
             # Use .get() with a default of {} to avoid KeyError if element_properties is incomplete
            props = [element_properties.get(el, {}).get(prop_name) for el in element_list if el in element_properties]
            props_dict[prop_name] = pd.Series(props) if props else pd.Series([np.nan]) # Store as Series for easy aggregation

        # Calculate aggregations
        results['num_unique_elements'] = len(set(element_list)) if element_list else 0 # Count unique elements

        for prop_name in ['atomic_number', 'atomic_mass', 'electronegativity', 'covalent_radius']:
            results[f'avg_{prop_name}'] = props_dict[prop_name].mean()
            results[f'min_{prop_name}'] = props_dict[prop_name].min()
            results[f'max_{prop_name}'] = props_dict[prop_name].max()

        # Add the 14th feature: std_electronegativity
        results['std_electronegativity'] = props_dict['electronegativity'].std() if len(props_dict['electronegativity'].dropna()) > 1 else 0.0 # Handle single element case


        return pd.Series(results)

    # Apply the function to create the new features
    aggregated_features = df['elements_parsed'].apply(calculate_elemental_aggregations)

    # Concatenate the new features to the original DataFrame
    df = pd.concat([df, aggregated_features], axis=1)

    print("Created 14 Host Material Features (num_unique_elements, avg/min/max for atomic number, atomic mass, electronegativity, covalent radius, and std_electronegativity).")

    df.drop(columns=['elements_parsed'], inplace=True) # Drop temporary column

else:
    print("Warning: 'elements' column not found. Skipping Host Material Feature engineering.")
```

## Intercalating Ion Features (Li, Na, Mg specific)

```python
# Predefined properties for common working ions
# Values are sourced from standard periodic table data.

ion_properties = {
    'Li': {'ionic_radius': 0.76, 'atomic_weight': 6.941, 'charge': 1},
    'Na': {'ionic_radius': 1.02, 'atomic_weight': 22.990, 'charge': 1},
    'Mg': {'ionic_radius': 0.72, 'atomic_weight': 24.305, 'charge': 2}
}
```

```python
# Intercalating Ion Features (5 features)

if element_properties:
    for ion_sym in allowed_ions:
        if ion_sym in element_properties:
            ion_properties[ion_sym]['atomic_number'] = element_properties[ion_sym].get('atomic_number')
            ion_properties[ion_sym]['electronegativity'] = element_properties[ion_sym].get('electronegativity')
        else:
            print(f"Warning: Properties for ion '{ion_sym}' not found in element_properties. Some ion features might be NaN.")
            # Set to NaN if not found to avoid KeyError
            ion_properties[ion_sym]['atomic_number'] = np.nan
            ion_properties[ion_sym]['electronegativity'] = np.nan

if 'working_ion' in df.columns:
    df['ionic_radius_ion'] = df['working_ion'].map(lambda x: ion_properties.get(x, {}).get('ionic_radius'))
    df['atomic_weight_ion'] = df['working_ion'].map(lambda x: ion_properties.get(x, {}).get('atomic_weight'))
    df['charge_ion'] = df['working_ion'].map(lambda x: ion_properties.get(x, {}).get('charge'))
    df['atomic_number_ion'] = df['working_ion'].map(lambda x: ion_properties.get(x, {}).get('atomic_number'))
    df['electronegativity_ion'] = df['working_ion'].map(lambda x: ion_properties.get(x, {}).get('electronegativity'))

    print("Created 5 Intercalating Ion Features (ionic_radius_ion, atomic_weight_ion, charge_ion, atomic_number_ion, electronegativity_ion).")
else:
    print("Warning: 'working_ion' column not found. Skipping Intercalating Ion Features.")
```

```python
# One-Hot Encoded categorical Features

if 'working_ion' in df.columns:
    # Define all categorical columns to be one-hot encoded
    columns_to_onehot_encode = ['working_ion']
    if 'crystal_system' in df.columns: # Add crystal_system
        columns_to_onehot_encode.append('crystal_system')

    if 'is_metal' in df.columns and df['is_metal'].dtype == 'bool':
        columns_to_onehot_encode.append('is_metal')
    if 'theoretical' in df.columns and df['theoretical'].dtype == 'bool':
        columns_to_onehot_encode.append('theoretical')

    print("\nApplying One-Hot Encoding to categorical features...")
    df = pd.get_dummies(df, columns=columns_to_onehot_encode, prefix=columns_to_onehot_encode, dummy_na=False, dtype=int)

    # Ensure specific 'working_ion' columns exist and create them if not
    # Prefix will be 'working_ion' as per get_dummies
    for ion in allowed_ions:
        col_name = f'working_ion_{ion}'
        if col_name not in df.columns:
            df[col_name] = 0 # Add column and fill with 0 if no samples of that ion were left after filter

    print(f"Created One-Hot Encoded Features for: {', '.join(columns_to_onehot_encode)}.")
    print(f"Shape after One-Hot Encoding: {df.shape}")
    print("Example of new encoded columns (first 5 rows):")
    # Adjust this print to show columns relevant to your encoded features
    print(df[[col for col in df.columns if any(p in col for p in ['working_ion_', 'crystal_system_', 'is_metal_', 'theoretical_'])]].head())

else:
    print("Warning: 'working_ion' column not found. Skipping One-Hot Encoded Ion Features.")
    # If crystal_system still needs encoding, handle it separately here
    if 'crystal_system' in df.columns:
        print("Warning: 'working_ion' not found, but encoding 'crystal_system' separately.")
        df = pd.get_dummies(df, columns=['crystal_system'], prefix='crystal_system', dummy_na=False, dtype=int)
        print(f"Shape after One-Hot Encoding for crystal_system: {df.shape}")
```

## Drop Redundant Columns

```python
# Drop original columns that have been transformed or replaced
columns_to_drop = ['elements']


if 'crystal_system' in df.columns:
    columns_to_drop.append('crystal_system')

if 'is_metal' in df.columns and df['is_metal'].dtype == 'bool':
    columns_to_drop.append('is_metal')
if 'theoretical' in df.columns and df['theoretical'].dtype == 'bool':
    columns_to_drop.append('theoretical')

df.drop(columns=[col for col in columns_to_drop if col in df.columns], errors='ignore', inplace=True)
print(f"\nDropped original columns: {columns_to_drop}")
```

```python
# Handle missing values for newly engineered features on df

newly_engineered_numeric_cols = [
    'ionic_radius_ion', 'atomic_weight_ion', 'charge_ion',
    'atomic_number_ion', 'electronegativity_ion',
    'e_fermi','nsites'
]

print("\nHandling missing values for newly engineered features and new MP features...")
for col in newly_engineered_numeric_cols:
    if col in df.columns:
        if df[col].isnull().any(): # Check if any NaN values are present in the column
            mean_val = df[col].mean()
            df[col] = df[col].fillna(mean_val)
            print(f"    Imputed NaN values in '{col}' with mean: {mean_val:.4f}")
        else:
            print(f"    '{col}' has no NaN values.")
    elif col not in df.columns: # For clarity if a listed column is entirely missing
        print(f"    Warning: Column '{col}' not found for NaN imputation.")
```

```python
# Final check for NaNs

nan_counts_final = df.isnull().sum()
nan_counts_final = nan_counts_final[nan_counts_final > 0]

if not nan_counts_final.empty:
    print("\nWarning: NaN values still present in DataFrame after imputation (before final dropna):")
    print(nan_counts_final)
else:
    print("\nNo NaN values detected in the DataFrame after imputation (before final dropna).")
```

```python
# Ensure original boolean columns are also explicitly cast to int
for col_prefix in ['is_metal', 'theoretical', 'working_ion']: # Check for both original name and new prefix for robustness
    # Check if the original boolean column still exists and is of boolean type
    if col_prefix in df.columns and df[col_prefix].dtype == 'bool':
        df[col_prefix] = df[col_prefix].astype(int)
        print(f"    Converted original '{col_prefix}' to int.")
    # Check if one-hot encoded versions are present (e.g., 'is_metal_0', 'is_metal_1')
    elif any(f"{col_prefix}_" in c for c in df.columns):
        print(f"    '{col_prefix}' or its related columns were one-hot encoded; no direct boolean conversion needed for original.")
```

```python
# Final dropna applied to ensure all *modeling-relevant* features and targets have data.

features_to_exclude_from_dropna_subset = [
    'material_id', 'formula_pretty',
    'battery_id', # If this column exists and should be excluded from dropna subset
    'framework_formula', # If this column exists and should be excluded from dropna subset
    'discharged_formula' # If this column exists and should be excluded from dropna subset
]
```

```python
# The subset for dropna should be all columns that are NOT excluded identifiers
dropna_subset = [col for col in df.columns if col not in features_to_exclude_from_dropna_subset]

initial_rows_before_final_dropna = len(df)
df.dropna(subset=dropna_subset, inplace=True) # Apply to df
print(f"Final dropna applied to ensure all *modeling-relevant* features and targets have data. New row count: {len(df)} (Dropped {initial_rows_before_final_dropna - len(df)} rows).")
```

```python
# Now df is your final, engineered, clean DataFrame. Create df_engineered as a copy if desired.
df_engineered = df.copy()

print(f"DataFrame after Feature Engineering has {len(df_engineered)} rows and {len(df_engineered.columns)} columns.")
print("New features created/transformed, and all final features/targets are non-null.")
print("\n--- First 5 rows of the DataFrame after Feature Engineering: ---")
print(df_engineered.head().to_markdown(index=False, numalign="left", stralign="left")) # Use df_engineered here
```

# Appendix: Detailed Feature Engineering Steps

This notebook performed a detailed feature engineering process on the battery materials data to prepare it for machine learning modeling. The objective was to transform the raw data into a richer set of predictive features by extracting and creating more meaningful numerical representations of both the host material and the intercalating ion, and to properly handle newly integrated materials properties.

Here's a breakdown of the key actions performed:

1.  **Data Loading**:
    - Loaded the raw battery materials data from the file `/content/drive/MyDrive/Colab Notebooks/Masters Research Project/Data/battery_materials_processed_v2.csv`.
    - The initial dataset contained 3709 rows and 25 columns.

2.  **Filter by Working Ion**:
    - Filtered the dataset to focus on common battery chemistries by keeping only entries where the `working_ion` is 'Li', 'Na', or 'Mg'.
    - This step dropped 264 rows, resulting in a DataFrame with 3445 rows.

3.  **Load Element Properties**:
    - Loaded comprehensive elemental properties from the PubChem CSV file located at `/content/drive/MyDrive/Colab Notebooks/Masters Research Project/Data/PubChemElements_all.csv`.
    - This data was used to create a lookup dictionary (`element_properties`) containing properties like atomic number, atomic mass, electronegativity, and covalent radius for each element.
    - Properties for 118 elements were successfully loaded. Note: 'AtomicRadius' from the CSV was used as a proxy for 'covalent_radius'.

4.  **Derived Host Material Features (14 Features)**:
    - For each material, features describing the host material's elemental composition were generated. This involved:
        - Parsing the string representation of the `elements` column into Python lists.
        - Calculating aggregate properties for the elements in each list using the loaded `element_properties` dictionary.
    - The 14 features created are:
        - `num_unique_elements`: The count of distinct elements in the material's formula.
        - Average, minimum, and maximum values for `atomic_number`.
        - Average, minimum, and maximum values for `atomic_mass`.
        - Average, minimum, and maximum values for `electronegativity`.
        - Average, minimum, and maximum values for `covalent_radius` (using AtomicRadius as a proxy).
        - `std_electronegativity`: The standard deviation of electronegativity for the elements.
    - A temporary column `elements_parsed` was used during calculation and then dropped.

5.  **Intercalating Ion Features (5 Features)**:
    - Properties specific to the `working_ion` (Li, Na, or Mg) were added to the DataFrame.
    - These features include:
        - `ionic_radius_ion`
        - `atomic_weight_ion`
        - `charge_ion`
        - `atomic_number_ion`
        - `electronegativity_ion`
    - These values were sourced from predefined properties (`ion_properties`) and supplemented with data from the `element_properties` dictionary.

6.  **One-Hot Encoded Categorical Features (14 Features)**:
    - Categorical columns were transformed into a numerical format suitable for machine learning models using one-hot encoding.
    - The columns encoded were:
        - `working_ion`: Resulting in `working_ion_Li`, `working_ion_Mg`, `working_ion_Na` (3 features).
        - `crystal_system`: Resulting in binary columns for each unique crystal system (e.g., `crystal_system_Cubic`, `crystal_system_Hexagonal`, etc.) (7 features).
        - `is_metal`: Resulting in `is_metal_False`, `is_metal_True` (2 features).
        - `theoretical`: Resulting in `theoretical_False`, `theoretical_True` (2 features).
    - One-hot encoding added a total of 14 new binary features to the DataFrame.
    - The shape of the DataFrame after one-hot encoding was (3445, 54).

7.  **Drop Redundant Columns**:
    - The original columns that were transformed or replaced by the feature engineering process were dropped to avoid redundancy and multicollinearity.
    - The dropped columns included: `elements`, `working_ion`, `crystal_system`, `is_metal`, and `theoretical`.

8.  **Handle Missing Values**:
    - Although the data from the Materials Project API was largely complete, a check and imputation step was performed for any potential missing values in the newly engineered numeric features and the newly added raw MP features (`e_fermi`, `nsites`).
    - Missing values in these specific columns were imputed using the mean of the respective column.
    - A final check confirmed that no NaN values were detected in the DataFrame after this imputation step.

9.  **Final Data Integrity Check**:
    - A final `dropna` operation was applied to the DataFrame, specifically targeting all columns *except* for identifier/descriptive columns (`material_id`, `formula_pretty`, `battery_type`).
    - This step ensured that every remaining data point had complete, non-null values for all features and target variables relevant for modeling.
    - 0 rows were dropped in this final step, confirming the completeness of the modeling-relevant data.

10. **Save Engineered Data**:
    - The final DataFrame (`df_engineered`) containing all the engineered and cleaned features was saved to a new CSV file: `/content/drive/MyDrive/Colab Notebooks/Masters Research Project/Data/battery_materials_engineered_features_v2.csv`.

**Outcome:**

The final DataFrame (`df_engineered`) has 3445 rows and 53 columns. These columns are categorized as:

-   **3 Identifier/Descriptive Columns**: `material_id`, `formula_pretty`, `battery_type`. These are not used as model inputs.
-   **3 Target Variables (Y)**: `theoretical_capacity_grav`, `average_voltage`, `energy_above_hull`. These are the properties to be predicted.
-   **47 Input Features (X)**: These are the features used to train the model and include:
    -   **14 Original Raw Features (retained/new from MP)**: `density`, `nelements`, `volume`, `band_gap`, `total_magnetization`, `theoretical_capacity_vol`, `energy_grav`, `energy_vol`, `num_steps`, `max_voltage_step`, `stability_charge`, `stability_discharge`, `e_fermi`, `nsites`.
    -   **14 Engineered Host Material Features**: Derived from elemental properties (e.g., `num_unique_elements`, `avg_atomic_number`, `max_electronegativity`, `std_electronegativity`, `min_covalent_radius`, etc.).
    -   **5 Engineered Intercalating Ion Features**: Properties of the working ion (`ionic_radius_ion`, `atomic_weight_ion`, `charge_ion`, `atomic_number_ion`, `electronegativity_ion`).
    -   **14 Engineered One-Hot Encoded Features**: Binary indicators for categorical properties (`working_ion_Li`, `working_ion_Mg`, `working_ion_Na`, `crystal_system_Cubic`, `crystal_system_Hexagonal`, `crystal_system_Monoclinic`, `crystal_system_Orthorhombic`, `crystal_system_Tetragonal`, `crystal_system_Triclinic`, `crystal_system_Trigonal`, `is_metal_False`, `is_metal_True`, `theoretical_False`, `theoretical_True`).

This comprehensive set of features is now ready for use in building and training machine learning models to predict battery material properties.
